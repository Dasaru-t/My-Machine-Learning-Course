<h1>Pandas: The Excel of Python</h1>
<p><b>Pandas</b> is the most important library for Data Science. If NumPy is the engine, Pandas is the dashboard.</p>
<p>It provides two key data structures:</p>
<ul>
  <li><b>Series:</b> A single column of data (like a 1D NumPy array with labels).</li>
  <li><b>DataFrame:</b> A table with rows and columns (like an Excel sheet or SQL table).</li>
</ul>
<p>In this tutorial, we will learn how to load, clean, and analyze a dataset efficiently.</p>
