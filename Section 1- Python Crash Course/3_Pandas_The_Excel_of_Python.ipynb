<h1>Pandas: The Excel of Python</h1>
<p><b>Pandas</b> is the most important library for Data Science. If NumPy is the engine, Pandas is the dashboard.</p>
<p>It provides two key data structures:</p>
<ul>
  <li><b>Series:</b> A single column of data (like a 1D NumPy array with labels).</li>
  <li><b>DataFrame:</b> A table with rows and columns (like an Excel sheet or SQL table).</li>
</ul>
<p>In this tutorial, we will learn how to load, clean, and analyze a dataset efficiently.</p>

```python
import pandas as pd
import numpy as np
from io import StringIO

# 1. Creating a DataFrame
# In real life, you usually load data from CSV, but let's create a mock dataset relevant to IT/Business
csv_data = """ID,Name,Role,Salary,Performance_Score
101,Krish,Data Scientist,120000,9.5
102,Sam,Data Analyst,85000,8.2
103,John,Manager,150000,7.8
104,Sara,Data Scientist,125000,9.1
105,Tom,Developer,90000,NaN
106,Emma,Developer,NaN,8.5
"""

# StringIO simulates a real file file for this example
df = pd.read_csv(StringIO(csv_data))

# 2. Inspecting Data
print("--- First 5 Rows ---")
print(df.head())

print("\n--- Data Info (Types & Nulls) ---")
print(df.info())

print("\n--- Statistical Summary ---")
# .describe() gives you quick stats (Count, Mean, Min, Max, Quartiles)
print(df.describe())
```
